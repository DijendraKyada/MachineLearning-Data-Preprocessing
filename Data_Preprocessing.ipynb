```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

# Task 1

___Reading into dfProcessed___

```python
dfProcessed=pd.read_csv('processed.cleveland.data',header=None)
dfProcessed.columns= ['Age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','health']
dfProcessed.head()
```

```python
dfProcessed.shape
```

___Reading into dfCleve___

```python
dfCleve=pd.read_csv('cleve.txt',skiprows=19,header = None,sep='\s+')
dfCleve.columns=['Age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','xx','health']
dfCleve=dfCleve.drop(['xx'], axis=1)
dfCleve.head()
```

```python
dfCleve.shape
```

# Task 2

```python
dfCleve['sex'][dfCleve['sex']=='male'] = 1
dfCleve['sex'][dfCleve['sex']=='fem'] = 0

dfCleve['cp'][dfCleve['cp']=='angina'] = 1
dfCleve['cp'][dfCleve['cp']=='abnang'] = 2
dfCleve['cp'][dfCleve['cp']=='notang'] = 3
dfCleve['cp'][dfCleve['cp']=='asympt'] = 4

dfCleve['fbs'][dfCleve['fbs']=='true'] = 1
dfCleve['fbs'][dfCleve['fbs']=='fal'] = 0

dfCleve['restecg'][dfCleve['restecg']=='norm'] = 0
dfCleve['restecg'][dfCleve['restecg']=='abn'] = 1
dfCleve['restecg'][dfCleve['restecg']=='hyp'] = 2


dfCleve['exang'][dfCleve['exang']=='true'] = 1
dfCleve['exang'][dfCleve['exang']=='fal'] = 0

dfCleve['slope'][dfCleve['slope']=='up'] = 1
dfCleve['slope'][dfCleve['slope']=='flat'] = 2
dfCleve['slope'][dfCleve['slope']=='down'] = 3

dfCleve['thal'][dfCleve['thal']=='norm'] = 3
dfCleve['thal'][dfCleve['thal']=='fix'] = 6
dfCleve['thal'][dfCleve['thal']=='rev'] = 7

dfCleve['health'][dfCleve['health']=='H'] = 0
dfCleve['health'][dfCleve['health']=='S1'] = 1
dfCleve['health'][dfCleve['health']=='S2'] = 2
dfCleve['health'][dfCleve['health']=='S3'] = 3
dfCleve['health'][dfCleve['health']=='S4'] = 4
```

```python
dfCleve.head()
```

```python
dfCleve=dfCleve[dfCleve['ca']!='?']
dfCleve=dfCleve[dfCleve['thal']!='?']
dfCleve.head()
```

```python
dfCP=pd.DataFrame(dfCleve,dtype=np.float)
```

```python
dfCP.head()
```

# Task 3

```python
dfCleve2=dfCleve.drop('health',axis=1)
```

```python
dfCP2=dfCP.drop('health',axis=1)
```

```python
dfCleve2.head()
```

```python
dfOnehot=pd.get_dummies(dfCleve2)
```

```python
dfOnehot.head()
```

```python
dfScaled = preprocessing.scale(dfOnehot)
```

```python
dfCleve['health']=dfCleve['health'].astype(int)
```

___Checking performance of original data___

```python
clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf, dfCP2, dfCleve['health'], cv=10)
scores 
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

___Checking performance of onehot data___

```python
clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf, dfOnehot, dfCleve['health'], cv=10)
scores 
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

___Checking performance of Scaled data___

```python
clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf, dfScaled, dfCleve['health'], cv=10)
scores 
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
